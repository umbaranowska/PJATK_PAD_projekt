```python
import pandas as pd
data = pd.read_csv('../02/clean_data02.csv', index_col='Unnamed: 0')
data.head(5)
```

```python
data['oceny'][3].split('\r\n')
```

```python
data[['srednia_ocena', 'x1', 'ocena_obsluga', 'x2', 'ocena_klimat', 'x3', 'trudnosc_wg_graczy', 'x4', 'x5']] = data['oceny'].str.split('\r\n', expand=True)
data.drop(columns=['oceny', 'x1', 'x2', 'x3', 'x4', 'x5'], inplace=True)
data
```

```python
data['srednia_ocena'] = data['srednia_ocena'].str.replace(' ', '')
data['ocena_obsluga'] = data['ocena_obsluga'].str.replace(',', '.')
data['ocena_obsluga'] = data['ocena_obsluga'].str.replace('/10', '')
data['ocena_klimat'] = data['ocena_klimat'].str.replace(',', '.')
data['ocena_klimat'] = data['ocena_klimat'].str.replace('/10', '')
```
