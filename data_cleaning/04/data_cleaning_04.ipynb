```python
import pandas as pd
data = pd.read_csv('../03/clean_data03.csv', index_col='Unnamed: 0')
data.head(5)
```

```python
data['liczba_graczy'] = data['liczba_graczy'].str.replace('osób|osoby', '')
data['czas_gry'] = data['czas_gry'].str.replace('minut|minuty', '')
data['liczba_ocen'] = data['liczba_ocen'].str.replace('opinii|opinie|opinia', '')
data['miejsce_w_polsce'] = data['miejsce_w_polsce'].str.replace('Polska: ', '')
data['miejsce_w_polsce'] = data['miejsce_w_polsce'].str.replace('. miejsce', '')
data
```

```python
data[['min_liczba_graczy', 'max_liczba_graczy']] = data['liczba_graczy'].str.split('-', expand=True)
data.drop(columns='liczba_graczy', inplace=True)
data
```

```python
data.to_csv('clean_data04.csv')
```
